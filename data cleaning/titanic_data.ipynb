```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=sns.load_dataset("titanic")
```

```python
data
```

```python
death=data[data.alive=='no']
```

```python
death
```

```python
live=data[data.alive=='yes']
```

```python
live
```

```python
death.sex.value_counts()
```

```python
live.sex.value_counts()
```

```python
plt.figure(figsize=(16,9))
plt.pie(death.sex.value_counts(),labels=death.sex.unique(),autopct='%i%%',colors=["r","g"],textprops={'fontsize':15})
plt.title('death_grf',fontsize=15)
```

```python
plt.figure(figsize=(15,9))
plt.pie(live.sex.value_counts(),labels=live.sex.unique(),autopct='%i%%',colors=['m','y'],textprops={'fontsize':20})
plt.title('Live_grf')
plt.show()
```

```python
plt.figure(figsize=(15,9))
plt.subplot(1,2,1)
plt.pie(death.sex.value_counts(),labels=death.sex.unique(),autopct='%i%%',colors=["r","g"],textprops={'fontsize':15})
plt.title('death_grf',fontsize=15)
plt.subplot(1,2,2)
plt.pie(live.sex.value_counts(),labels=live.sex.unique(),autopct='%i%%',colors=['m','y'],textprops={'fontsize':20})
plt.title('Live_grf')
```
